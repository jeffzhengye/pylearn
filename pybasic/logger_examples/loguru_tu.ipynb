```python
from loguru import logger
import sys

# 设置日志同时输出到控制台和日志文件，并解决缓冲问题
logger.add(sys.stdout, format="{time} {level} {message}", enqueue=True)
logger.add("output.log", rotation="10 MB", catch=True, backtrace=True, diagnose=True)

@logger.catch()
def a(n):
    return 1 / n

ret = a(0)

print('i am back, with', ret)
```
