## tags

* Django模板语言中的标签（Tag）：
在Django模板语言中，标签是一种特殊类型的指令，它告诉模板引擎执行某种操作。标签通常用一对{% %}包围，比如{% if %}、{% for %}等。这些标签可以控制流程（如条件语句和循环），加载外部的标签库，包含其他模板片段，或者执行其他特定任务。
* 自定义模板标签（Custom Template Tags）：
Django允许开发者创建自己的模板标签，这些自定义标签可以封装复杂的逻辑或数据库查询，使得模板代码更简洁、更易于维护。自定义模板标签通常是通过定义一个Python函数并将其装饰为register.simple_tag或register.inclusion_tag来实现的。
