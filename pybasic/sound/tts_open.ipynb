```python
from fairseq.checkpoint_utils import load_model_ensemble_and_task_from_hf_hub
from fairseq.models.text_to_speech.hub_interface import TTSHubInterface
import IPython.display as ipd


models, cfg, task = load_model_ensemble_and_task_from_hf_hub(
    "facebook/tts_transformer-zh-cv7_css10",
    arg_overrides={"vocoder": "hifigan", "fp16": False}
)
# print(type(models), models)
model = models[0]
models[0] = model.cuda()
TTSHubInterface.update_cfg_with_data_cfg(cfg, task.data_cfg)
generator = task.build_generator(models, cfg)

text = "您好，这是试运行。ipad为什么只输出两个字啊"

sample = TTSHubInterface.get_model_input(task, text)
print(sample)
wav, rate = TTSHubInterface.get_prediction(task, model, generator, sample)

ipd.Audio(wav, rate=rate)
```

```python
text = "您好，这是试运行。how much is an ipad为什么只输出两个字啊"

sample = TTSHubInterface.get_model_input(task, text)
net_input = sample['net_input']
import torch

for k in net_input.keys():
    v = net_input[k]
    if isinstance(v, torch.Tensor):
        net_input[k] = v.cuda()

sample['speaker'] = sample['speaker'].cuda()
print(sample)
wav, rate = TTSHubInterface.get_prediction(task, model, generator, sample)

ipd.Audio(wav.cpu().numpy(), rate=rate)
```

```python
class Registry:
    def __init__(self):
        self._registry = {}

    def register(self, name):
        print(f"register {name}")
        def decorator(cls):
            print(f"decorator")
            self._registry[name] = cls()
            return cls
        return decorator

    def get(self, name):
        return self._registry.get(name)


# 示例使用
# 创建注册表实例
registry = Registry()

# 注册装饰器
@registry.register("class1")
class MyClass1:
    def __init__(self):
        print("init MyClass1")
    
    def say_hello(self):
        print("Hello from MyClass1")

@registry.register("class2")
class MyClass2:
    def say_hello(self):
        print("Hello from MyClass2")

# 从注册表获取对象并使用
# obj1 = registry.get("class1")
# if obj1:
#     obj1.say_hello()

# obj2 = registry.get("class2")
# if obj2:
#     obj2.say_hello()
```
